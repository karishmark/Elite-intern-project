```python
# import the libiaries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the dataset
df_diamond = pd.read_csv("diamonds.csv")
df_diamond
```

```python
df_diamond.info()
```

```python
#checking the missing values
df_diamond.isnull().sum()
```

```python
#statistics numerical columns 
df_diamond.describe()
```

```python
#check unique values
df_diamond["cut"].unique()
```

```python
#data visualzation 
sns.set_style("whitegrid")
```

```python
#distribuation of diamond price
plt.figure(figsize= (4,4))
sns.histplot(df_diamond["price"], bins=45, kde= True, color= "red")
plt.title("DISTRIBUATION OF DIAMOND PRICE")
plt.xlabel("price")
plt.ylabel("count")
plt.show()
```

```python
# boxplot of carat vs price
plt.figure(figsize= (4,4))
sns.boxplot(x = "carat", y = "price", data = df_diamond, palette="coolwarm")
plt.xlabel("carat weight")
plt.ylabel("price of the diamond")
plt.title("CARAT VS PRICE")
plt.xticks(rotation = 45)
plt.show()
```

```python
#scatter plot cut vs carat
plt.figure(figsize=(4,4))
sns.scatterplot(x = "cut", y = "carat", data = df_diamond, alpha = 0.5, color = "purple")
plt.title("CUT VS carat")
plt.xlabel("cut")
plt.ylabel("carat")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

categorical_columns = ["cut", "color", "clarity"]
for col in categorical_columns:
    df_diamond[col] = label_encoder.fit_transform(df_diamond[col])
    
plt.figure(figsize=(8,4))
sns.heatmap(df_diamond.corr(), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("FEATURE OF CORRELATION MATRIX")
plt.show()
```
